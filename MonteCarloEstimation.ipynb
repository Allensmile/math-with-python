# Monte Carlo Estimation

Let's draw a circle by randomly generating points in the space $x$,$y$ such that $x^{2}+y^{2}=1$

```python
import random

NUM_POINTS = 10000

x = [random.uniform(-1,1) for i in xrange(NUM_POINTS)]
y = [random.uniform(-1,1) for i in xrange(NUM_POINTS)]

circle_x = []
circle_y = []

outsiders_x = []
outsiders_y = []

for i in xrange(NUM_POINTS):
    if x[i]**2 + y[i]**2 <= 1:
        circle_x.append(x[i])
        circle_y.append(y[i])
    else:
        outsiders_x.append(x[i])
        outsiders_y.append(y[i])

fig = plt.figure()
fig.set_size_inches(10, 10)
_ = scatter(outsiders_x, outsiders_y, s=1, color='green')
_ = scatter(circle_x, circle_y, s=1, color='red')
```

# Now let's try to estimate $\pi$

Here's a few facts:

* The area of a circle is $\pi r^2$.
* Since our circle only has a radius of 1, its area is $\pi$.
* The area of the square is $2*2 = 4$.
* The ratio of the **area of the circle** to the **area of the square** should be the same as the ratio of the **points inside the circle** to the **points inside the square**
* So the ratio of $\frac{\pi}{4} = \frac{circle\ points}{total\ points}$.  So $\pi = 4 * \frac{circle\ points}{total\ points}$

```python
print "Estimate of area of circle (pi):", 4 * (len(circle_x)*1.0 / len(x))
```

# Estimate area under a curve

Estimate area under curve: $y = 5 * \sin(6*x) + \sin(2*x)$ from 0 to 10

```python
f = lambda x: 5 * sin(6 * x) + 3 * sin(2 * x)

x = linspace(0, 10, 1000)
y = f(x)

_ = plot(x,y)
```

```python
rand_x = lambda: random.uniform(0, 10)
rand_y = lambda: random.uniform(-8, 8)

xs = [rand_x() for i in xrange(NUM_POINTS)]
ys = [rand_y() for i in xrange(NUM_POINTS)]

under_x = []
under_y = []
over_x = []
over_y = []

for i in xrange(NUM_POINTS):
    if ys[i] <= f(xs[i]):
        under_x.append(xs[i])
        under_y.append(ys[i])
    else:
        over_x.append(xs[i])
        over_y.append(ys[i])

fig = plt.figure()
fig.set_size_inches(12, 8)
_ = scatter(under_x, under_y, s=1, color='red')
_ = scatter(over_x, over_y, s=1, color='green')
```

```python
# Area = area of domain rectangle * num_points_under/num_points_total
area = 10*16 * len(under_x)*1.0/len(xs)
print "Estimate of area under the curve:", area

# Sanity check: it should be around 80 (if it ended on an even period, it would be 80).
```
