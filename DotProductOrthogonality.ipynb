## Dot Product and Orthogonality

$$\vec{A} \cdot \vec{B} = \sum\limits_{i=1}^n A_i B_i $$

Conceptually, the dot product can be thought of as a measure of similarity between 2 vectors.  Here's how I think of the value of the dot product:

* Large **positive** values mean the vectors "agree" - they are very similar
    - This is used for finding the similarity between 2 vectors in the [Cross-correlation](http://en.wikipedia.org/wiki/Cross-correlation) algorithm.
* Large **negative** values mean the vectors "disagree"
* Value of (or near) **zero** means they don't even give answers to the same questions - they are orthogonal.
    - Let's explore orthogonality more below.


### Orthogonality

Let's think more about what it means for vectors to be orthogonal.  Geometrically, you can think of orthogonal vector being at a "right angle".  For instance, (1,0) and (0,1) form a right angle if you plot them:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines

fig, ax = plt.subplots()

fig.set_size_inches(6,6)                  # Make graph square
_ = plt.scatter([-0.1,1.1],[-0.1,1.1],s=0.01) # Move graph window a little left and down

ax.arrow(0, 0, 1, 0, head_width=0.05, head_length=0.1, fc='k', ec='k')
ax.arrow(0, 0, 0, 1, head_width=0.05, head_length=0.1, fc='k', ec='k')

plt.show()
```

### Dot product and vector projection

The dot product is used in the calculations for [vector projection](http://en.wikipedia.org/wiki/Vector_projection).  When thinking about a vector being projected on to another, it makes sense that projecting one vector onto another which is orthogonal to it would result in a value of 0.  Imagine a light shining directly down from above the vertical arrow in the graph above - it would create no shadow.  But if the vectors were pointing in almost the same direction, the shadow would almost cover the bottom vector.  And if the top vector were at an angle almost 180 degrees from the bottom vector, the shadow would look like an arrow pointing in the opposite direction of the bottom vector.

## Waves of different frequencies are orthoganal

**Ya!  It's pretty crazy - waves of different frequencies are orthogonal to each other**.  This means you can actually think of different frequencies as different dimensions.

Consider the 2 waves, one has a frequency of 1Hz, and the othe has a frequency of 2Hz.

```python
x = np.linspace(0, 2, 1000)    # In seconds
wave1 = np.sin(1 * 2*np.pi*x)     # 1Hz wave
wave2 = np.sin(2 * 2*np.pi*x)     # 2Hz wave

_ = plt.plot(x, wave1, 'blue', x, wave2, 'red')
```

Let's now look at each of these waves as vectors (the vectors are simply values of the waves at various points in time) and take their dot product; it should come out to about 0.

```python
np.dot(wave1, wave2)
```

As you can see, the value is very near **zero**.  This phenomenon is related to how the [Fourier Transform](http://en.wikipedia.org/wiki/Fourier_transform) can find the power in a given wavelength by multiplying the whole complex wave by that wavelength.
